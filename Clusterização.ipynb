Carregando os arquivos xlsx da pasta de scrapping para gerar os .txts e fazer a clusterização

```python
import os
import pandas as pd
import numpy as np

# Caminho para a pasta que contém os arquivos .xlsx
path = r'C:\Users\mathe\OneDrive\Área de Trabalho\SoccerIA\webScrappingOddspedia'

# Lista todos os arquivos na pasta
files = os.listdir(path)

# Filtra a lista de arquivos para incluir apenas os arquivos .xlsx
xlsx_files = [f for f in files if f.endswith('.xlsx')]

dfss = []

colunas_interesse = [
    'team1_goals', 'team2_goals',
    'team1_total_shots', 'team2_total_shots', 'team1_shots_on_target',
    'team2_shots_on_target', 'team1_fouls', 'team2_fouls', 'team1_corners',
    'team2_corners','team1_red_cards', 'team2_red_cards', 'team1_shots_out','team2_shots_out'
]

# Carrega cada arquivo .xlsx em um dataframe e armazena na lista dfss
for file in xlsx_files:
    data_path = os.path.join(path, file)
    df = pd.read_excel(data_path)

    # Remove as linhas que possuem NaN nas colunas de interesse
    df.dropna(subset=colunas_interesse, inplace=True)

    # Calcula a média para as colunas de interesse
    medias = df[colunas_interesse].mean()

    # Cria o dicionário
    dict_medias = medias.to_dict()

    # Salva o dicionário em um arquivo .txt
    nome_txt = os.path.splitext(file)[0] + '.txt'  # Substitui a extensão por .txt
    caminho_txt = os.path.join(path, nome_txt)
    
    with open(caminho_txt, 'w') as arquivo_txt:
        for key, value in dict_medias.items():
            linha = f"{key}: {value}\n"
            arquivo_txt.write(linha)
    
    dfss.append(df)
    print(file)
    print(df.shape)

# Combine all the DataFrames into a single DataFrame
combined_df_13c = pd.concat(dfss, ignore_index=True)
print(combined_df_13c.shape)
print(combined_df_13c.columns)
```

Importando os txts da pasta 'Datasets finais europeus xlsx'

```python
import os

# Caminho para a pasta que contém os arquivos .txt
path = r'C:\Users\mathe\OneDrive\Área de Trabalho\SoccerIA\Datasets finais europeus xlsx'

# Lista todos os arquivos na pasta
files = os.listdir(path)

# Filtra a lista de arquivos para incluir apenas os arquivos .txt
txt_files = [f for f in files if f.endswith('.txt')]

# Dicionário para armazenar os dicionários de cada arquivo
all_data = {}

# Processa cada arquivo .txt
for file in txt_files:
    with open(os.path.join(path, file), 'r') as f:
        # Lê as linhas do arquivo
        lines = f.readlines()
        
        # Cria um dicionário para armazenar os dados do arquivo atual
        data_dict = {}
        
        # Processa cada linha para extrair a chave e o valor
        for line in lines:
            key, value = line.strip().split(': ')
            data_dict[key] = round(float(value), 3)  # Arredonda o valor para 2 casas decimais
        
        # Adiciona o dicionário ao dicionário geral com o nome do arquivo como chave (sem a extensão .txt)
        all_data[file.replace('.txt', '')] = data_dict

# Agora, o dicionário 'all_data' contém todos os dados dos arquivos .txt
print(all_data)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np



# Feature scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Aplicando DBSCAN
dbscan = DBSCAN(eps=2.5, min_samples=2)
clusters = dbscan.fit_predict(scaled_data)

# Adicionando a coluna de clusters ao DataFrame original
df['cluster'] = clusters

# Ordena o DataFrame 'df' pela coluna 'cluster' e salva as alterações de volta para 'df'
df.sort_values(by='cluster', inplace=True)

# Cria um escritor de arquivos Excel
with pd.ExcelWriter('clusters_output.xlsx') as writer:
    # Para cada cluster, salva em uma aba diferente
    for cluster, group in df.groupby('cluster'):
        # Reseta o índice, se 'group' for o índice do DataFrame
        group.reset_index(inplace=True)
        
        # Nome da aba
        sheet_name = f'Cluster {cluster}'
        # Salva o grupo de dados (cluster) na aba correspondente
        group.to_excel(writer, sheet_name=sheet_name, index=False)
        
        # Para visualização no console, conforme o formato que você gostou
        print(f"Cluster {cluster}:")
        print(group)
        print("\n" + "-"*50 + "\n")


# Observações:
# - `eps`: É a distância máxima entre dois pontos de amostra para que um seja considerado na vizinhança do outro.
# - `min_samples`: Número mínimo de pontos de amostra em uma vizinhança para que um ponto seja considerado como um ponto central.

```

Busca de vizinhos próximos

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler


# Feature scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Configura o algoritmo de vizinhos mais próximos
n_neighbors = 10  # Número de vizinhos que queremos encontrar; isso pode ser ajustado
model_knn = NearestNeighbors(n_neighbors=n_neighbors, algorithm='auto', metric='euclidean')

# Ajusta o modelo
model_knn.fit(scaled_data)

# Para cada competição, encontra as k competições mais próximas
# Isso retorna duas matrizes: 
# - 'distances' contém as distâncias até os vizinhos mais próximos
# - 'indices' contém os índices desses vizinhos no DataFrame original
distances, indices = model_knn.kneighbors(scaled_data)

# Para cada competição (índice), lista as competições mais próximas e suas distâncias
for i, idx in enumerate(indices):
    print(f"Competição {df.index[i]} tem as seguintes competições vizinhas:")
    for j in range(1, len(idx)):
        print(f" - Competição {df.index[idx[j]]} com distância {distances[i][j]}")
    print("\n" + "-"*50 + "\n")
```
